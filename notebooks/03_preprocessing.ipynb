# 03. Data Preprocessing & Feature Engineering

```
import pandas as pd
import pickle
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```
users = pd.read_csv('../data/users.csv')
products = pd.read_csv('../data/products.csv')
events = pd.read_csv('../data/user_events.csv')
```

```
# Handle missing values
users['age'].fillna(users['age'].median(), inplace=True)
users['preferred_category'].fillna('Unknown', inplace=True)
```

```
# Feature Engineering
encoder = LabelEncoder()
users['gender_encoded'] = encoder.fit_transform(users['gender'].astype(str))
scaler = MinMaxScaler()
users['age_scaled'] = scaler.fit_transform(users[['age']])
```

```
# Save preprocessors
with open('../models/encoder.pkl', 'wb') as f:
    pickle.dump(encoder, f)
with open('../models/scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```
